## 1. Introduction

Matplotlib is a python library to plot graphs. We will mostly use pyplot package of matplotlib library. See the code below. For plotting a histogram (frequency vs value) we use matplotlib library, and for visualising data also. In this workbook 2D plots are covered, for 3D plots, there can be a seperate workbook.

```python
import matplotlib.pyplot as plt
import numpy as np
```

## 2. plot

The plot takes mainly two arguments x-coordinates and y-coordinates. It plots the curve joining these points. The curve can be line also.

```python
xpoints = np.array([1,5])
ypoints = np.array([1, 6])
plt.plot(xpoints, ypoints)
```

```python
xpoints = np.linspace(0,10,100)
ypoints = np.sin(xpoints)
plt.plot(xpoints,ypoints)
```

You can plot values vs index by the following code.

```python
plt.plot([3, 5, 1, 0, -4, 8])
```

## 3. labelling

If you see above graphs, they are not lebelled. You can add titles for X and Y axis, title for graph and much more. See the code below.

```python
plt.plot(xpoints, np.sin(xpoints), marker = '+', color = 'red', label = 'sine')
plt.plot(xpoints, np.cos(xpoints), marker = '.', color = 'blue', label = 'cosine')

plt.xlabel('Points') # X axis label
plt.ylabel('Sine values') # Y axis label
plt.title('Graph for trigonometric functions') # Title of graph
plt.legend() # Legend adds notation for each curve
plt.show() # Shows the graph at last
```

## 4. Subplots

You can have multiple plots in one figure. See the code below.

```python
xpoints = np.linspace(0,10,100)

plt.subplot(1,2,1) # (rows, cols, plot no.)
plt.plot(xpoints, np.sin(xpoints))

plt.subplot(1,2,2)
plt.plot(xpoints, np.cos(xpoints))

plt.show()
```

## Bar Plot

```python
subjects = ['English','Maths','Science','History','Computer']
marks = [70, 50, 53, 40, 100] # Good practice to define data first

plt.bar(subjects, marks, color='red',alpha=0.4, width=0.5) #Alpha - color intensity
plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.title('Subjects vs marks')
plt.show()
```

```python
subjects = ['English','Maths','Science','History','Computer']
mid_marks = [70, 50, 53, 40, 100]
end_marks = [30, 80, 50, 60, 100]


plt.bar(subjects, mid_marks, label='Mid Exam Marks', color='blue', alpha=0.7)
plt.bar(subjects, end_marks, label='End Exam Marks', color='red', alpha=0.7)

plt.xlabel('Subjects')
plt.ylabel('Marks')
plt.legend() #Legend to show information
plt.title('Subjects vs Exam Marks')

plt.show()
```

```python
subjects = ['English','Maths','Science','History','Computer']
mid_marks = [70, 50, 53, 40, 100]
end_marks = [30, 80, 50, 60, 100]

barWidth = 0.25 # Width of bar
 
# Set position of bar on X axis
br1 = np.arange(len(mid_marks))
br2 = [x + barWidth for x in br1]
 
# Make the plot
plt.bar(br1, mid_marks, color ='r', width = barWidth, label ='Mid Exam')
plt.bar(br2, end_marks, color ='g', width = barWidth, label ='End Exam')
 
# Adding Xticks
plt.xlabel('Subjects', fontweight ='bold', fontsize = 15)
plt.ylabel('Marks', fontweight ='bold', fontsize = 15)
plt.xticks([r + barWidth for r in range(len(mid_marks))], subjects)
plt.title("Exam Marks")
 
plt.legend()
plt.show()
```

## Pie Chart

```python
subjects = ['English','Maths','Science','History','Computer']
marks = [70, 50, 53, 40, 100] # Good practice to define data first

plt.pie(marks, labels=subjects, autopct='%1.2f%%') #autopct = 1i - integer and no decimal, 1.2f - float and 2 decimals
plt.title("Result")
plt.show()
```

## Scatter Plot

```python
xpoints = np.linspace(0,10,50)

plt.scatter(xpoints,np.sin(xpoints), color='g', s = 10)
plt.scatter(xpoints,np.cos(xpoints), color='b', marker='^', s=10)
plt.title("Plot of Sine and Cosine")
plt.xlabel("Angle")
plt.ylabel("Value")
plt.show()
```

## 8. Histogram

The most important plot and you will often use it in ML. It is the distribution of values (frequency vs value).

```python
mean = 0
std = 1
n = 500

x = np.random.normal(mean, std, n)
plt.hist(x)
plt.show()
```
